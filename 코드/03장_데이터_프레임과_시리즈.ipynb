#### 3.1.4. 매개변수와 인수, 기본값


```python
# 매개변수와 인수, 기본값을 확인하기 위한 함수 정의

def func(x=0, y=1, z=2):
    return x + (2 * y) + (3 * z)

func(y=3, x=1, z=2) # 매개변수와 인수를 함께 입력하면 순서에 관계없이 입력 가능하다.
func(1, 2, 3) # x=1, y=2, z=3으로 순서에 맞추어 입력하면 매개변수는 생략이 가능하다
func(1, y=3, z=2) # 파이썬에서는 주로 일부 매개변수만 생략한다
func(y=3) # y=3이 입력되고, 나머지는 x=0, z=2의 기본값이 적용
func(2) # x=2가 입력되고, 나머지는 y=1, z=2의 기본값이 적용
```
