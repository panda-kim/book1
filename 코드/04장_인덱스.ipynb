### 4.1. 인덱싱과 슬라이싱

#### 4.1.1. 대괄호 인덱싱

````python
# 코드 4-1. 실습을 위한 데이터 프레임 생성
import pandas as pd
data1 = [[84, 74, 86, 83, 60, 78],
         [69, 60, 91, 60, 91, 76],
         [88, 89, 80, 62, 62, 60],
         [72, 87, 90, 61, 98, 99],
         [90, 88, 64, 62, 83, 69]]
df = pd.DataFrame(data1, index=['A', 'B', 'C', 'D', 'E'],
                  columns=['국어','영어','수학','사회','과학','불어'])
# 원본인 df는 두고 df를 복제한 df1를 생성 실습한다.
df1 = df.copy()
df1
````

````python
# 코드 4-2. 대괄호 인덱싱으로 국어 열을 시리즈로 추출
df1['국어']
````

````python
# 코드 4-3. 대괄호 인덱싱으로 복수의 열을 추출하기
df1[['영어', '과학']]
````

````python
# 코드 4-4. 동일한 열을 여러 번 인덱싱
df1[['과학', '영어', '과학']]
````

````python
# 코드 4-5. 대괄호 인덱싱으로 열의 순서 변경하기
df1[['과학', '영어', '수학', '사회', '불어', '국어']]
````

대괄호에 존재하지 않는 열 이름을 입력해 출력하면 KeyError가 발생한다.

```python
df1['체육']
```
```python
KeyError: '체육'
```
df1에는 체육 열이 없으므로 KeyError가 발생한다

````python
# 코드 4-6. 대괄호 인덱싱으로 열 생성하기
df1['체육'] = 'pass'
df1
````

````python
# 코드 4-7. 서로 다른 값을 부여해 열 생성하기
df1['도덕'] = [95, 87, 83, 91, 77]
df1
````

````python
# 코드 4-8. 복수의 열 생성하기
df1[['미술', '음악']] = 90
df1
````

#### 4.1.3. 대괄호 인덱싱으로 열 수정하기

````python
# 코드 4-9. 열 수정하기
df1['국어'] = 99
df1
````

#### 4.1.4. 대괄호 슬라이싱

````python
# 코드 4-10. df에서 B 학생부터 D 학생까지 추출하기
df['B':'D']
````

#### 4.1.5. 키 인덱싱과 슬라이싱(loc 인덱서)

````python
# 코드 4-11. df를 복제한 df2에서 A 학생의 국어 성적 추출하기
df2 = df.copy()
df2.loc['A', '국어']
````

````python
# 코드 4-12. loc 인덱서로 복수의 행과 열을 인덱싱
df2.loc[['A', 'C', 'D'], ['국어', '수학']]
````

````python
# 코드 4-13. B 행부터 D 행의 영어부터 과학 성적까지 추출하기
df2.loc['B':'D', '영어':'과학']
````

````python
# 코드 4-14. 시작점이 처음이거나 끝 지점이 마지막이면 생략 가능
df2.loc['A':'D', '수학':'불어'] # 모두 기입한 코드
df2.loc[:'D', '수학':] # 생략도 가능
````

````python
# 코드 4-15. 모든 행을 가져온다면 행의 슬라이싱은 콜론(:)으로 대체 가능
df2.loc['A':'E', '수학':'과학'] # 모두 기입한 코드
df2.loc[:, '수학':'과학'] # 콜론으로 대체고 가능
````

````python
# 코드 4-16. 모든 열을 가져온다면 열의 슬라이싱은 생략 가능
df2.loc[:'D', :] # 열의 슬라이싱을 콜론으로 대체도 가능하지만
df2.loc[:'D'] # 모든 열을 가져올 때 열의 슬라이싱 전체를 생략 가능
````

행을 전부 가져올 경우라도 행의 슬라이싱은 생략될 수 없다.
코드 4-15처럼 사용해야 한다.

```python
df2.loc['수학':'과학'] # X
df2.loc[:, '수학':'과학'] # O
```

````python
# 코드 4-17. 인덱싱과 슬라이싱은 행과 열에 각각 적용 가능
df2.loc[:'D', ['과학', '수학']]
````

#### 4.1.6. loc 인덱서로 행과 열 생성하기

````python
# 코드 4-18. B와 D에게만 80을 부여한 실험 열 생성하기
df2.loc[['B', 'D'], '실험'] = 80 # B와 D는 80, 나머지는 NaN인 실험 열 생성
df2
````

````python
# 코드 4-19. A의 과학 점수를 70점으로 수정하기
df2.loc['A', '과학'] = 70
df2
````

````python
# 코드 4-20. 전 과목 점수가 60인 F학생의 데이터 생성하기
df2.loc['F'] = 60
df2
````

#### 4.1.7. 로케이션 인덱싱과 슬라이싱(iloc 인덱서)

````python
# 코드 4-21. df를 복제한 df3로 A 학생의 국어 점수 로케이션 인덱싱
df3 = df.copy()
df3.iloc[0, 0]
````

````python
# 코드 4-22. A, C, D 학생의 국어와 수학 성적을 가져오기
df3.iloc[[0, 2, 3], [0, 2]]
````

````python
# 코드 4-23. 행을 1:3, 열을 1:4로 슬라이싱
df3.iloc[1:3, 1:4]
````

````python
# 코드 4-24. 행은 B부터 D, 열은 영어부터 과학까지 로케이션 슬라이싱
df3.iloc[1:4, 1:5]
````

````python
# 코드 4-25. 행은 A부터 D, 열은 수학부터 불어까지 로케이션 슬라이싱
df3.iloc[:4, 2:]
````

````python
# 코드 4-26. 첫 행을 시리즈로 추출하기
df3.iloc[0]
````

````python
# 코드 4-27. 마지막 행을 시리즈로 추출하기
df3.iloc[-1] # 음의 로케이션
````

````python
# 코드 4-28. 인덱싱과 슬라이싱은 행과 열에 각각 적용 가능
df3.iloc[1:4, [-2, 3, -2]]
````
