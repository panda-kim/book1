### 5.2. 시리즈의 연산

#### 5.2.1. 시리즈와 단일 값의 연산

```python
# 코드 5-1. 실습 예제 코드
import pandas as pd
s1 = pd.Series([1, 5, 0, 4], index=list('ABCD'))
s2 = pd.Series([1, 2, 3, 3], index=list('ABCD'))
s3 = pd.Series([3, 4, 2, 3], index=list('ABDE'))
```

```python
# 코드 5-2. 시리즈와 단일 값의 산술 연산 덧셈
s1 + 2 # 연산자를 사용하거나
s1.add(2) # 연산 함수를 사용할 수 있다.
```

```python
# 코드 5-3. 시리즈와 단일 값의 비교 연산
s1 > 3 # 연산자
s1.gt(3) # 연산 함수
```

#### 5.2.2. 시리즈 간의 연산

```python
# 코드 5-4. 인덱스가 일치하는 두 시리즈의 뺄셈
s1 - s2 # 연산자
s1.sub(s2) # 연산 함수
```

```python
# 코드 5-5. 인덱스가 일치하는 두 시리즈의 동등 비교 연산(==)
s1 == s2 # 연산자
s1.eq(s2) # 연산 함수
```

```python
# 코드 5-6. 시리즈의 논리 연산
(s1 > 0) & (s1 < 3)  # 연산자를 사용하면 반드시 소괄호로 묶어야 한다.
s1.gt(0) & s1.lt(3) # 연산 함수
```

```python
# 코드 5-7. 인덱스의 구성이 다른 두 시리즈의 곱셈
s1 * s3 # 연산자
s1.mul(s3) # 연산 함수
```

```python
# 코드 5-8. 인덱스의 구성이 다른 두 시리즈의 비교 연산
s1.le(s3) # 연산 함수를 사용해야 한다.
```

#### 5.2.2. 데이터 프레임에서 열 간의 연산

```python
# 코드 5-9. 실습 예제 코드
import pandas as pd
data1 = {'성': ['김', '강', '권', '박'],
         '이름': ['판다', '승주', '보아', '효신'],
         '국어': [79, 65, 92, 95],
         '영어': [68, 83, 87, 79]}
df = pd.DataFrame(data1)
df
```

```python
# 코드 5-10. 국어 열과 영어 열을 더해 총점 열 생성하기
df['총점'] = df['국어'] + df['영어']
df
```

```python
# 코드 5-11. 총점 열을 2로 나누어 평균 열 생성하기
df['평균'] = df['총점'] / 2
df
```

```python
# 코드 5-12. 국어 점수가 영어 점수보다 높은 사람 확인하기
df['국어'] > df['영어']
```

```python
# 코드 5-13. 평균 점수가 80점 미만인 사람 확인하기
df['평균'] < 80
```

```python
# 코드 5-14. 국어 점수가 영어 점수보다 높으면서, 80점 미만인 사람 확인하기
(df['국어'] > df['영어']) & (df['평균'] < 80)
```

```python
# 코드 5-15. 성과 공백, 이름을 합쳐 성명 열을 생성
df['성명'] = df['성'] + ' ' + df['이름']
df[['성명', '국어', '영어', '평균']] # 성명, 국어, 영어, 평균 열만 출력
```

### 5.3. 데이터 프레임의 연산


#### 5.3.1. 데이터 프레임과 단일 값의 연산

```python
# 코드 5-16. 실습 예제 코드
import pandas as pd
data1 = [[0, 1, 1], [0, 2, 1], [4, 2, 5], [2, 2, 5]]
data2 = [[3, 2, 1], [3, 4, 5], [0, 3, 4], [0, 0, 4]]
df1 = pd.DataFrame(data1, index=list('abcd'), columns=list('ABC'))
df2 = pd.DataFrame(data2, index=list('abcd'), columns=list('ABC'))
df3 = pd.DataFrame(data2, index=list('abde'), columns=list('ABD'))
```

```python
# 코드 5-17. 데이터 프레임과 단일 값의 산술 연산(몫)
df1 // 3 # 연산자
df1.floordiv(3) # 연산 함수
```

```python
# 코드 5-18. 데이터 프레임과 단일 값의 비교 연산(다르다)
df1 != 2 # 연산자
df1.ne(2) # 연산 함수
```

#### 5.3.2. 데이터 프레임 간의 연산


```python
# 코드 5-19. 인덱스와 컬럼즈가 일치하는 두 데이터 프레임의 산술 연산
df1 + df2 # 연산자
df1.add(df2) # 연산 함수
```

```python
# 코드 5-20. 인덱스와 컬럼즈가 일치하는 두 데이터 프레임의 비교 연산
df1 >= df2 # 연산자
df1.ge(df2) # 연산 함수
```

```python
# 코드 5-21. 인덱스와 컬럼즈가 일치하지 않는 두 데이터 프레임의 산술 연산
df1 + df3 # 연산자
df1.add(df3) # 연산 함수
```

#### 5.3.3. NaN을 대체해 연산하기(연산 함수의 fill_value)

```python
# 코드 5-22. fill_value=0을 입력해 코드 5-21을 수행
df1.add(df3, fill_value=0)
```

#### 5.3.4. 데이터 프레임과 시리즈의 연산(브로드 캐스팅)

```python
# 코드 5-23. 실습 예제 코드
import pandas as pd
data1 = [[1, 1, 1], [1, 2, 3], [4, 3, 5], [2, 2, 4]]
df = pd.DataFrame(data1, index=list('abcd'), columns=list('ABC'))
s1 = pd.Series([0, 1, 2], index=list('ABC'))
s2 = pd.Series([0, 1, 2, 3], index=list('abcd'))
```

```python
# 코드 5-24. 데이터 프레임과 시리즈의 곱셈(브로드 캐스팅)
df * s1 # 연산자
df.mul(s1) # 연산 함수
```

```python
# 코드 5-25. 브로드 캐스팅의 축 지정
df.mul(s2, axis=0) # axis를 0으로 지정하려면 연산 함수를 사용해야 한다.
```

```python
# 코드 5-26. 브로드 캐스팅의 축방향을 정확히 지정하지 않았을 때
df * s2
```

### 엑셀 예제 3. 다양한 통계 함수 실습하기

```python
# 코드 5-40. 엑셀 파일 첫 번째 시트에서 데이터 프레임 불러오기
import pandas as pd
url2 = 'https://github.com/panda-kim/book1/blob/main/07statics.xlsx?raw=true'
df1 = pd.read_excel(url2)
df1
```

```python
# 코드 5-41. 조 배정하기
df1['조'] = df1['구분'].eq('조장').cumsum()
df1
```

```python
# 코드 5-42. 엑셀 파일 두 번째 시트에서 데이터 프레임 불러오기
df2 = pd.read_excel(url2, sheet_name=1)
df2
```

```python
# 코드 5-43. 각 팀의 승수 구하기
df2.eq('Win').sum()
```

```python
# 코드 5-44. 각 팀의 승률 구하기
df2.eq('Win').mean()
```

```python
# 코드 5-45. 엑셀 파일 세 번째 시트에서 데이터 프레임 불러오기
df3 = pd.read_excel(url2, sheet_name=2)
df3
```

```python
# 코드 5-46. 각 과목의 평균 구하기
df3.loc[:, '국어':'과학'].mean()
```

```python
# 코드 5-47. 각 과목의 표준편차 구하기
df3.loc[:, '국어':'과학'].std()
```

```python
# 코드 5-48. 브로드 캐스팅을 위한 시리즈 생성하기
s = pd.Series([40, 40, 40, 40, 60], index=df3.columns)
s
```

```python
# 코드 5-49. 모든 과목이 40점 이상이고 평균이 60점 이상인 합격자 수 구하기
df3.ge(s).all(axis=1).sum()
```

```python
# 코드 5-50. 엑셀 파일 네 번째 시트에서 데이터 프레임 불러오기
df4 = pd.read_excel(url2, sheet_name=3, index_col=0)
df4
```

```python
# 코드 5-51. 각 사람의 출근 일수 구하기
df4.eq('출근').sum(axis=1)
```

```python
# 코드 5-52. 각 사람의 첫 출근일 구하기
df4.eq('출근').idxmax(axis=1)
```

```python
# 코드 5-53. 각 사람의 마지막 출근일 구하기
df4.eq('출근').cumsum(axis=1).idxmax(axis=1)
```
