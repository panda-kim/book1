### 5.2. ÏãúÎ¶¨Ï¶àÏùò Ïó∞ÏÇ∞

#### 5.2.1. ÏãúÎ¶¨Ï¶àÏôÄ Îã®Ïùº Í∞íÏùò Ïó∞ÏÇ∞

```python
# ÏΩîÎìú 5-1. Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
s1 = pd.Series([1, 5, 0, 4], index=list('ABCD'))
s2 = pd.Series([1, 2, 3, 3], index=list('ABCD'))
s3 = pd.Series([3, 4, 2, 3], index=list('ABDE'))
```

```python
# ÏΩîÎìú 5-2. ÏãúÎ¶¨Ï¶àÏôÄ Îã®Ïùº Í∞íÏùò ÏÇ∞Ïà† Ïó∞ÏÇ∞ ÎçßÏÖà
s1 + 2 # Ïó∞ÏÇ∞ÏûêÎ•º ÏÇ¨Ïö©ÌïòÍ±∞ÎÇò
s1.add(2) # Ïó∞ÏÇ∞ Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎã§.
```

```python
# ÏΩîÎìú 5-3. ÏãúÎ¶¨Ï¶àÏôÄ Îã®Ïùº Í∞íÏùò ÎπÑÍµê Ïó∞ÏÇ∞
s1 > 3 # Ïó∞ÏÇ∞Ïûê
s1.gt(3) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

#### 5.2.2. ÏãúÎ¶¨Ï¶à Í∞ÑÏùò Ïó∞ÏÇ∞

```python
# ÏΩîÎìú 5-4. Ïù∏Îç±Ïä§Í∞Ä ÏùºÏπòÌïòÎäî Îëê ÏãúÎ¶¨Ï¶àÏùò Î∫ÑÏÖà
s1 - s2 # Ïó∞ÏÇ∞Ïûê
s1.sub(s2) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

```python
# ÏΩîÎìú 5-5. Ïù∏Îç±Ïä§Í∞Ä ÏùºÏπòÌïòÎäî Îëê ÏãúÎ¶¨Ï¶àÏùò ÎèôÎì± ÎπÑÍµê Ïó∞ÏÇ∞(==)
s1 == s2 # Ïó∞ÏÇ∞Ïûê
s1.eq(s2) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

```python
# ÏΩîÎìú 5-6. ÏãúÎ¶¨Ï¶àÏùò ÎÖºÎ¶¨ Ïó∞ÏÇ∞
(s1 > 0) & (s1 < 3)  # Ïó∞ÏÇ∞ÏûêÎ•º ÏÇ¨Ïö©ÌïòÎ©¥ Î∞òÎìúÏãú ÏÜåÍ¥ÑÌò∏Î°ú Î¨∂Ïñ¥Ïïº ÌïúÎã§.
s1.gt(0) & s1.lt(3) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

```python
# ÏΩîÎìú 5-7. Ïù∏Îç±Ïä§Ïùò Íµ¨ÏÑ±Ïù¥ Îã§Î•∏ Îëê ÏãúÎ¶¨Ï¶àÏùò Í≥±ÏÖà
s1 * s3 # Ïó∞ÏÇ∞Ïûê
s1.mul(s3) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

```python
# ÏΩîÎìú 5-8. Ïù∏Îç±Ïä§Ïùò Íµ¨ÏÑ±Ïù¥ Îã§Î•∏ Îëê ÏãúÎ¶¨Ï¶àÏùò ÎπÑÍµê Ïó∞ÏÇ∞
s1.le(s3) # Ïó∞ÏÇ∞ Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥Ïïº ÌïúÎã§.
```

#### 5.2.2. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóêÏÑú Ïó¥ Í∞ÑÏùò Ïó∞ÏÇ∞

```python
# ÏΩîÎìú 5-9. Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
data1 = {'ÏÑ±': ['ÍπÄ', 'Í∞ï', 'Í∂å', 'Î∞ï'],
         'Ïù¥Î¶Ñ': ['ÌåêÎã§', 'ÏäπÏ£º', 'Î≥¥ÏïÑ', 'Ìö®Ïã†'],
         'Íµ≠Ïñ¥': [79, 65, 92, 95],
         'ÏòÅÏñ¥': [68, 83, 87, 79]}
df = pd.DataFrame(data1)
df
```

```python
# ÏΩîÎìú 5-10. Íµ≠Ïñ¥ Ïó¥Í≥º ÏòÅÏñ¥ Ïó¥ÏùÑ ÎçîÌï¥ Ï¥ùÏ†ê Ïó¥ ÏÉùÏÑ±ÌïòÍ∏∞
df['Ï¥ùÏ†ê'] = df['Íµ≠Ïñ¥'] + df['ÏòÅÏñ¥']
df
```

```python
# ÏΩîÎìú 5-11. Ï¥ùÏ†ê Ïó¥ÏùÑ 2Î°ú ÎÇòÎàÑÏñ¥ ÌèâÍ∑† Ïó¥ ÏÉùÏÑ±ÌïòÍ∏∞
df['ÌèâÍ∑†'] = df['Ï¥ùÏ†ê'] / 2
df
```

```python
# ÏΩîÎìú 5-12. Íµ≠Ïñ¥ Ï†êÏàòÍ∞Ä ÏòÅÏñ¥ Ï†êÏàòÎ≥¥Îã§ ÎÜíÏùÄ ÏÇ¨Îûå ÌôïÏù∏ÌïòÍ∏∞
df['Íµ≠Ïñ¥'] > df['ÏòÅÏñ¥']
```

```python
# ÏΩîÎìú 5-13. ÌèâÍ∑† Ï†êÏàòÍ∞Ä 80Ï†ê ÎØ∏ÎßåÏù∏ ÏÇ¨Îûå ÌôïÏù∏ÌïòÍ∏∞
df['ÌèâÍ∑†'] < 80
```

```python
# ÏΩîÎìú 5-14. Íµ≠Ïñ¥ Ï†êÏàòÍ∞Ä ÏòÅÏñ¥ Ï†êÏàòÎ≥¥Îã§ ÎÜíÏúºÎ©¥ÏÑú, 80Ï†ê ÎØ∏ÎßåÏù∏ ÏÇ¨Îûå ÌôïÏù∏ÌïòÍ∏∞
(df['Íµ≠Ïñ¥'] > df['ÏòÅÏñ¥']) & (df['ÌèâÍ∑†'] < 80)
```

```python
# ÏΩîÎìú 5-15. ÏÑ±Í≥º Í≥µÎ∞±, Ïù¥Î¶ÑÏùÑ Ìï©Ï≥ê ÏÑ±Î™Ö Ïó¥ÏùÑ ÏÉùÏÑ±
df['ÏÑ±Î™Ö'] = df['ÏÑ±'] + ' ' + df['Ïù¥Î¶Ñ']
df[['ÏÑ±Î™Ö', 'Íµ≠Ïñ¥', 'ÏòÅÏñ¥', 'ÌèâÍ∑†']] # ÏÑ±Î™Ö, Íµ≠Ïñ¥, ÏòÅÏñ¥, ÌèâÍ∑† Ïó¥Îßå Ï∂úÎ†•
```

### 5.3. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùò Ïó∞ÏÇ∞


#### 5.3.1. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÍ≥º Îã®Ïùº Í∞íÏùò Ïó∞ÏÇ∞

```python
# ÏΩîÎìú 5-16. Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
data1 = [[0, 1, 1], [0, 2, 1], [4, 2, 5], [2, 2, 5]]
data2 = [[3, 2, 1], [3, 4, 5], [0, 3, 4], [0, 0, 4]]
df1 = pd.DataFrame(data1, index=list('abcd'), columns=list('ABC'))
df2 = pd.DataFrame(data2, index=list('abcd'), columns=list('ABC'))
df3 = pd.DataFrame(data2, index=list('abde'), columns=list('ABD'))
```

```python
# ÏΩîÎìú 5-17. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÍ≥º Îã®Ïùº Í∞íÏùò ÏÇ∞Ïà† Ïó∞ÏÇ∞(Î™´)
df1 // 3 # Ïó∞ÏÇ∞Ïûê
df1.floordiv(3) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

```python
# ÏΩîÎìú 5-18. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÍ≥º Îã®Ïùº Í∞íÏùò ÎπÑÍµê Ïó∞ÏÇ∞(Îã§Î•¥Îã§)
df1 != 2 # Ïó∞ÏÇ∞Ïûê
df1.ne(2) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

#### 5.3.2. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ Í∞ÑÏùò Ïó∞ÏÇ∞


```python
# ÏΩîÎìú 5-19. Ïù∏Îç±Ïä§ÏôÄ Ïª¨ÎüºÏ¶àÍ∞Ä ÏùºÏπòÌïòÎäî Îëê Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùò ÏÇ∞Ïà† Ïó∞ÏÇ∞
df1 + df2 # Ïó∞ÏÇ∞Ïûê
df1.add(df2) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

```python
# ÏΩîÎìú 5-20. Ïù∏Îç±Ïä§ÏôÄ Ïª¨ÎüºÏ¶àÍ∞Ä ÏùºÏπòÌïòÎäî Îëê Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùò ÎπÑÍµê Ïó∞ÏÇ∞
df1 >= df2 # Ïó∞ÏÇ∞Ïûê
df1.ge(df2) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

```python
# ÏΩîÎìú 5-21. Ïù∏Îç±Ïä§ÏôÄ Ïª¨ÎüºÏ¶àÍ∞Ä ÏùºÏπòÌïòÏßÄ ÏïäÎäî Îëê Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùò ÏÇ∞Ïà† Ïó∞ÏÇ∞
df1 + df3 # Ïó∞ÏÇ∞Ïûê
df1.add(df3) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

#### 5.3.3. NaNÏùÑ ÎåÄÏ≤¥Ìï¥ Ïó∞ÏÇ∞ÌïòÍ∏∞(Ïó∞ÏÇ∞ Ìï®ÏàòÏùò fill_value)

```python
# ÏΩîÎìú 5-22. fill_value=0ÏùÑ ÏûÖÎ†•Ìï¥ ÏΩîÎìú 5-21ÏùÑ ÏàòÌñâ
df1.add(df3, fill_value=0)
```

#### 5.3.4. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÍ≥º ÏãúÎ¶¨Ï¶àÏùò Ïó∞ÏÇ∞(Î∏åÎ°úÎìú Ï∫êÏä§ÌåÖ)

```python
# ÏΩîÎìú 5-23. Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
data1 = [[1, 1, 1], [1, 2, 3], [4, 3, 5], [2, 2, 4]]
df = pd.DataFrame(data1, index=list('abcd'), columns=list('ABC'))
s1 = pd.Series([0, 1, 2], index=list('ABC'))
s2 = pd.Series([0, 1, 2, 3], index=list('abcd'))
```

```python
# ÏΩîÎìú 5-24. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÍ≥º ÏãúÎ¶¨Ï¶àÏùò Í≥±ÏÖà(Î∏åÎ°úÎìú Ï∫êÏä§ÌåÖ)
df * s1 # Ïó∞ÏÇ∞Ïûê
df.mul(s1) # Ïó∞ÏÇ∞ Ìï®Ïàò
```

```python
# ÏΩîÎìú 5-25. Î∏åÎ°úÎìú Ï∫êÏä§ÌåÖÏùò Ï∂ï ÏßÄÏ†ï
df.mul(s2, axis=0) # axisÎ•º 0ÏúºÎ°ú ÏßÄÏ†ïÌïòÎ†§Î©¥ Ïó∞ÏÇ∞ Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥Ïïº ÌïúÎã§.
```

```python
# ÏΩîÎìú 5-26. Î∏åÎ°úÎìú Ï∫êÏä§ÌåÖÏùò Ï∂ïÎ∞©Ìñ•ÏùÑ Ï†ïÌôïÌûà ÏßÄÏ†ïÌïòÏßÄ ÏïäÏïòÏùÑ Îïå
df * s2
```
