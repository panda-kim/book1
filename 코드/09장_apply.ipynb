#### 9.2.2. ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ìï®ÏàòÏôÄ apply

````python
# ÏΩîÎìú 9-6. ÌååÏù¥Ïç¨ÏóêÏÑú Î¨∏ÏûêÏó¥ÏùÑ Ïó≠ÏàúÏúºÎ°ú Î≥ÄÌôòÌïòÎäî Î∞©Î≤ï
x0 = 'abc'
x0[::-1]
````

````python
# ÏΩîÎìú 9-7. Î¨∏ÏûêÏó¥ÏùÑ Ïó≠ÏàúÏúºÎ°ú Î≥ÄÌôòÌïòÎäî ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ìï®Ïàò rev1 ÏÉùÏÑ±ÌïòÍ∏∞
def rev1(x):
    return x[::-1]
````

````python
# ÏΩîÎìú 9-8. Ìï®Ïàò rev1Ïóê Î≥ÄÏàò x0Î•º ÏûÖÎ†•Ìï¥ ÌôïÏù∏ÌïòÍ∏∞
rev1(x0)
````

````python
# ÏΩîÎìú 9-9. dfÏùò ÎãµÏïà Ïó¥Ïùò Î¨∏ÏûêÏó¥ÏùÑ Ïó≠ÏàúÏúºÎ°ú Î≥ÄÌôòÌïú ÏãúÎ¶¨Ï¶à ÏÉùÏÑ±ÌïòÍ∏∞
df['ÎãµÏïà'].apply(rev1)
````

#### 9.2.3. lambda Ìï®ÏàòÏôÄ apply

````python
# ÏΩîÎìú 9-10. defÎ°ú rev1ÏùÑ Ï†ïÏùòÌïòÍ≥† applyÎ•º Ï†ÅÏö©Ìïú ÏΩîÎìú Ï†ÑÏ≤¥ Î™®Ïùå
def rev1(x):
    return x[::-1]
df['ÎãµÏïà'].apply(rev1)
````

````python
# ÏΩîÎìú 9-11. lambda Ìï®ÏàòÎ°ú dfÏùò ÎãµÏïà Ïó¥Ïùò Î¨∏ÏûêÏó¥ÏùÑ Ïó≠ÏàúÏúºÎ°ú Î≥ÄÌôòÌïú ÏãúÎ¶¨Ï¶à ÏÉùÏÑ±
df['ÎãµÏïà'].apply(lambda x: x[::-1])
````

````python
# ÏΩîÎìú 9-12. ÏΩîÎìú 9-11Ïùò Í≤∞Í≥ºÎ°ú Ïó≠Ïàú Ïó¥ÏùÑ ÏÉùÏÑ±ÌïòÍ∏∞
df['Ïó≠Ïàú'] = df['ÎãµÏïà'].apply(lambda x: x[::-1])
df
````

````python
# ÏΩîÎìú 9-13. ÎÑ§ Í∏ÄÏûê Ïù¥ÏÉÅÏùò ÌöåÎ¨∏ÏùÑ ÎãµÏïàÏúºÎ°ú ÏûëÏÑ±Ìïú ÌïôÏÉù Ï∞æÍ∏∞
cond1 = df['Í∏ÄÏûêÏàò'] >= 4
cond2 = df['ÎãµÏïà'] == df['Ïó≠Ïàú']
df[cond1 & cond2]
````

#### 9.2.4. ÏãúÎ¶¨Ï¶àÏóê apply Ìï®Ïàò Ï†ÅÏö©Ìï† Îïå Ïú†ÏùòÌï† Ï†ê

````python
# ÏΩîÎìú 9-14. Î≤°ÌÑ∞Ìôî Ïó∞ÏÇ∞Ïù¥ Í∞ÄÎä•Ìï† ÎïåÎäî apply Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÎäî Í≤ÉÏù¥ ÎÇ´Îã§.
s = pd.Series([1, 2, 3, 4])
s + 1 # Î≤°ÌÑ∞Ìôî Ïó∞ÏÇ∞: Í∂åÏû• ÏΩîÎìú
s.apply(lambda x: x + 1) # apply Ï†ÅÏö©: Í∂åÏû•ÌïòÏßÄ ÏïäÎäî ÏΩîÎìú
````

#### 9.2.5. lambda Ìï®Ïàò Ï∂îÍ∞Ä ÌïôÏäµ


````python
# ÏΩîÎìú 9-15. lambda Ìï®Ïàò ÏÉùÏÑ± Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
s = pd.Series(['cat01', 'dog02', 'bird03'])
s
````

````python
# ÏΩîÎìú 9-16. Í∞Å ÏÖÄÎßàÎã§ Ìï®ÏàòÎ•º Ï†ÅÏö©Ìï† Í≤ÉÏù¥Îãà ÌïòÎÇòÏùò ÏÖÄÏùÑ xÎ°ú ÏßÄÏ†ï
x = s[0]
x
````

````python
# ÏΩîÎìú 9-17. xÏùò ÎßàÏßÄÎßâ Îëê Î¨∏ÏûêÎ•º Ï†úÍ±∞ÌïòÎäî ÏΩîÎìú ÏÉùÏÑ±
x[:-2]
````

````python
# ÏΩîÎìú 9-18. ÏΩîÎìú 9-17ÏùÑ lambda Ìï®ÏàòÏóê ÏûÖÎ†•Ìï¥ apply Ìï®ÏàòÏóê Ï†ÅÏö©
s.apply(lambda x: x[:-2])
````

````python
# ÏΩîÎìú 9-19. sÏùò Î¨∏ÏûêÏó¥ Í∏∏Ïù¥Í∞Ä 5ÏôÄ Í∞ôÎ©¥ 'O' ÏïÑÎãàÎ©¥ 'X'Î•º Î∞òÌôò
s.apply(lambda x: 'O' if len(x) ==  5 else 'X')
````

### 9.3. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê apply Ìï®Ïàò Ï†ÅÏö©ÌïòÍ∏∞

#### 9.3.1. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê apply Ìï®Ïàò Ï†ÅÏö©ÌïòÍ∏∞

````python
# ÏΩîÎìú 9-23. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê apply Ìï®Ïàò Ï†ÅÏö© Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
data = {'Íµ≠Ïñ¥': {'A': '87', 'B': '70', 'C': '98', 'D': '-'},
        'ÏòÅÏñ¥': {'A': '-', 'B': '72', 'C': '76', 'D': '77'},
        'ÏàòÌïô': {'A': '70', 'B': '73', 'C': '-', 'D': '95'}}
df = pd.DataFrame(data)
df
````

#### 9.3.2. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê apply Ìï®Ïàò Ï†ÅÏö©Ìï† Îïå Ï∂ï ÏßÄÏ†ï

````python
# ÏΩîÎìú 9-26. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóê apply Ìï®Ïàò Ï∂ï ÏßÄÏ†ï Ï†ÅÏö© Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
data1 = {'Íµ≠Ïñ¥': {'A': 90, 'B': 70, 'C': 80, 'D': float('nan')},
         'ÏòÅÏñ¥': {'A': float('nan'), 'B': 90, 'C': 80, 'D': 90},
         'ÏàòÌïô': {'A': 20, 'B': 50, 'C': float('nan'), 'D': 40}}
df1 = pd.DataFrame(data1)
df1
````

````python
# ÏΩîÎìú 9-27. lambda Ìï®ÏàòÎ•º ÏâΩÍ≤å ÏÉùÏÑ±ÌïòÍ∏∞ ÏúÑÌï¥ df1Ïùò Íµ≠Ïñ¥ Ïó¥ÏùÑ xÎ°ú ÏßÄÏ†ï
x = df1['Íµ≠Ïñ¥']
x
````

````python
# ÏΩîÎìú 9-28. ÏãúÎ¶¨Ï¶à xÏùò ÏµúÏÜüÍ∞íÏúºÎ°ú xÏùò Í≤∞Ï∏°Í∞íÏùÑ ÎåÄÏ≤¥ÌïòÎäî ÏΩîÎìú
x.fillna(x.min())
````

````python
# ÏΩîÎìú 9-29. Í∞Å Í≥ºÎ™©(Í∞Å Ïó¥)Ïùò ÏµúÏÜüÍ∞íÏúºÎ°ú Í≤∞Ï∏°Í∞í ÎåÄÏ≤¥ÌïòÍ∏∞
df1.apply(lambda x: x.fillna(x.min()))
````

````python
# ÏΩîÎìú 9-30. Í∞Å Í∞úÏù∏(Í∞Å Ìñâ)Ïùò ÏµúÏÜüÍ∞íÏúºÎ°ú Í≤∞Ï∏°Í∞í ÎåÄÏ≤¥ÌïòÍ∏∞
df1.apply(lambda x: x.fillna(x.min()), axis=1)
````

#### 9.3.3. Î≥µÏàò Ïó¥Ïùò Îç∞Ïù¥ÌÑ∞Î•º ÏûÖÎ†•ÌïòÎäî lambda Ìï®ÏàòÎ•º Í∞Å ÌñâÏóê Ï†ÅÏö©ÌïòÍ∏∞

````python
# ÏΩîÎìú 9-31. haversine ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò
!pip install haversine
````

````python
# ÏΩîÎìú 9-32. ÏúÑÎèÑÏôÄ Í≤ΩÎèÑÍ∞Ä ÏïåÎ†§ÏßÑ Îëê ÏßÄÏ†êÏùò ÏßÅÏÑ†Í±∞Î¶¨ Íµ¨ÌïòÍ∏∞
import haversine
seoul = [37.51, 127.05] # ÏúÑÎèÑ, Í≤ΩÎèÑ
pusan = [35.12, 129.08]
haversine.haversine(seoul, pusan, unit='km')
````

````python
# ÏΩîÎìú 9-33. Í∞Å ÎèÑÏãúÏôÄ ÏÑúÏö∏Ïùò ÏßÅÏÑ†Í±∞Î¶¨Î•º Íµ¨Ìï† Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
data2 = {'ÎèÑÏãú': ['ÏÑúÏö∏', 'Î∂ÄÏÇ∞', 'Í¥ëÏ£º', 'ÎåÄÏ†Ñ', 'Í∞ïÎ¶â', 'ÎåÄÍµ¨'],
         'ÏúÑÎèÑ': [37.51, 35.12, 35.13, 36.35, 37.75, 35.87],
         'Í≤ΩÎèÑ': [127.05, 129.08, 126.9, 127.42, 128.88, 128.6]}
df2 = pd.DataFrame(data2)
df2
````

````python
# ÏΩîÎìú 9-34. lambda Ìï®Ïàò ÏÉùÏÑ±ÏùÑ ÏúÑÌï¥ df2Ïùò Îëê Î≤àÏß∏ ÌñâÏùÑ Î≥ÄÏàò xÎ°ú ÏßÄÏ†ïÌïòÍ∏∞
x = df2.iloc[1]
x
````

````python
# ÏΩîÎìú 9-35. Î≥ÄÏàò xÏôÄ Î≥ÄÏàò seoul, Ìï®Ïàò haversineÏúºÎ°ú ÏÑúÏö∏, Î∂ÄÏÇ∞ Í±∞Î¶¨ Íµ¨ÌïòÍ∏∞
haversine.haversine(seoul, [x['ÏúÑÎèÑ'], x['Í≤ΩÎèÑ']], unit='km')
````

````python
# ÏΩîÎìú 9-36. df2ÏóêÏÑú Í∞Å ÎèÑÏãúÏôÄ ÏÑúÏö∏Í≥ºÏùò Í±∞Î¶¨ Íµ¨ÌïòÍ∏∞
df2['Í±∞Î¶¨(ÏÑúÏö∏)'] = df2.apply(
    lambda x: haversine.haversine(seoul, [x['ÏúÑÎèÑ'], x['Í≤ΩÎèÑ']], unit='km'),
    axis=1
)
df2
````
