### 9.2. 시리즈에 apply 함수 적용하기

#### 9.2.1. 시리즈에 apply 함수 적용하기

````python
# 코드 9-1. 네 글자 이상의 회문을 답하라는 문제에 대한 각 학생의 답안 데이터
import pandas as pd
data = {'이름': ['강승주', '송중기', '권보아', '김판다'],
        '답안': ['mom', 'get', 'pandas', 'level']}
df = pd.DataFrame(data)
df
````

````python
# 코드 9-2. len 함수 확인하기
len('python')
````

````python
# 코드 9-3. df의 답안 열에 len 함수 적용하기
len(df['답안'])
````

````python
# 코드 9-4. df의 답안 열에 apply로 len 함수 적용하기
df['답안'].apply(len)
````

````python
# 코드 9-5. 코드 9-4의 결과를 df 글자수 열로 생성하기
df['글자수'] = df['답안'].apply(len)
df
````

#### 9.2.2. 사용자 정의 함수와 apply

````python
# 코드 9-6. 파이썬에서 문자열을 역순으로 변환하는 방법
x0 = 'abc'
x0[::-1]
````

````python
# 코드 9-7. 문자열을 역순으로 변환하는 사용자 정의 함수 rev1 생성하기
def rev1(x):
    return x[::-1]
````

````python
# 코드 9-8. 함수 rev1에 변수 x0를 입력해 확인하기
rev1(x0)
````

````python
# 코드 9-9. df의 답안 열의 문자열을 역순으로 변환한 시리즈 생성하기
df['답안'].apply(rev1)
````

#### 9.2.3. lambda 함수와 apply

````python
# 코드 9-10. def로 rev1을 정의하고 apply를 적용한 코드 전체 모음
def rev1(x):
    return x[::-1]
df['답안'].apply(rev1)
````

````python
# 코드 9-11. lambda 함수로 df의 답안 열의 문자열을 역순으로 변환한 시리즈 생성
df['답안'].apply(lambda x: x[::-1])
````

````python
# 코드 9-12. 코드 9-11 결과를 역순 열로 생성하기
df['역순'] = df['답안'].apply(lambda x: x[::-1])
df
````

````python
# 코드 9-13. 네 글자 이상의 회문을 답안으로 작성한 학생 찾기
cond1 = df['글자수'] >= 4
cond2 = df['답안'] == df['역순']
df[cond1 & cond2]
````

#### 9.2.4. 시리즈에 apply 함수 적용할 때 유의할 점

````python
# 코드 9-14. 벡터화 연산이 가능할 때는 apply 함수를 사용하지 않는 것이 낫다.
s = pd.Series([1, 2, 3, 4])
s + 1 # 벡터화 연산: 권장 코드
s.apply(lambda x: x + 1) # apply 적용: 권장하지 않는 코드
````

#### 9.2.5. lambda 함수 추가 학습


````python
# 코드 9-15. lambda 함수 생성 실습 예제 코드
s = pd.Series(['cat01', 'dog02', 'bird03'])
s
````

````python
# 코드 9-16. 각 셀마다 함수를 적용할 것이니 하나의 셀을 x로 지정
x = s[0]
x
````

````python
# 코드 9-17. x의 마지막 두 문자를 제거하는 코드 생성
x[:-2]
````

````python
# 코드 9-18. 코드 9-17을 lambda 함수에 넣고 apply 적용
s.apply(lambda x: x[:-2])
````

````python
# 코드 9-19. s의 문자열 길이가 5와 같면 'O' 아니면 'X'를 반환
s.apply(lambda x: 'O' if len(x) ==  5 else 'X')
````

### 9.3. 데이터 프레임에 apply 함수 적용하기

````python
# 코드 9-23. 데이터 프레임에 apply 함수 적용 실습 예제 코드
import pandas as pd
data = {'국어': {'A': '87', 'B': '70', 'C': '98', 'D': '-'},
        '영어': {'A': '-', 'B': '72', 'C': '76', 'D': '77'},
        '수학': {'A': '70', 'B': '73', 'C': '-', 'D': '95'}}
df = pd.DataFrame(data)
df
````

#### 9.3.2. 데이터 프레임에 apply 함수 적용할 때 축 지정

````python
# 코드 9-26. 데이터 프레임에 apply 함수 축 지정 적용 실습 예제 코드
data1 = {'국어': {'A': 90, 'B': 70, 'C': 80, 'D': float('nan')},
         '영어': {'A': float('nan'), 'B': 90, 'C': 80, 'D': 90},
         '수학': {'A': 20, 'B': 50, 'C': float('nan'), 'D': 40}}
df1 = pd.DataFrame(data1)
df1
````

````python
# 코드 9-27. lambda 함수를 쉽게 생성하기 위해 df1의 국어 열을 x로 지정
x = df1['국어']
x
````

````python
# 코드 9-28. 시리즈 x의 최솟값으로 x의 결측값을 대체하는 코드
x.fillna(x.min())
````

````python
# 코드 9-29. 각 과목(각 열)의 최솟값으로 결측값 대체기
df1.apply(lambda x: x.fillna(x.min()))
````

````python
# 코드 9-30. 각 개인(각 행)의 최솟값으로 결측값 채우기
df1.apply(lambda x: x.fillna(x.min()), axis=1)
````

#### 9.3.3. 복수 열의 데이터를 입력하는 lambda 함수를 각 행에 적용하기

````python
# 코드 9-31. haversine 라이브러리 설치
!pip install haversine
````

````python
# 코드 9-32. 위도와 경도가 알려진 두 지점의 직선거리 구하기
import haversine
seoul = [37.51, 127.05] # 위도, 경도
pusan = [35.12, 129.08]
haversine.haversine(seoul, pusan, unit='km')
````

````python
# 코드 9-33. 각 도시와 서울의 직선거리를 구할 실습 예제 코드
data2 = {'도시': ['서울', '부산', '광주', '대전', '강릉', '대구'],
         '위도': [37.51, 35.12, 35.13, 36.35, 37.75, 35.87],
         '경도': [127.05, 129.08, 126.9, 127.42, 128.88, 128.6]}
df2 = pd.DataFrame(data2)
df2
````

````python
# 코드 9-34. lambda 함수 생성을 위해 df2의 두 번째 행을 변수 x로 지정하기
x = df2.iloc[1]
x
````

````python
# 코드 9-35. 변수 x와 변수 seoul, 함수 haversine으로 서울, 부산 거리 구하기
haversine.haversine(seoul, [x['위도'], x['경도']], unit='km')
````

````python
# 코드 9-36. df2에서 각 도시와 서울과의 거리 구하기
df2['거리(서울)'] = df2.apply(
    lambda x: haversine.haversine(seoul, [x['위도'], x['경도']], unit='km'),
    axis=1
)
df2
````
